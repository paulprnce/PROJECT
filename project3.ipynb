```python
x=a.drop(['Ethanol_gL'],axis=1)
y=a['Ethanol_gL']
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
categorical_cols=['Raw_Material', 'Strain', 'Reactor_Type', 'Pretreatment']
numerical_cols=['Lignin_percent', 'Moisture_percent', 'Ethanol_Tolerance_percent',
       'Temperature_C', 'pH', 'Fermentation_hours', 'Agitation_rpm',
       'Aeration_vvm', 'Inoculum_percent', 'Enzyme_gL',
       'Total_Sugar_gL', 'Process_Efficiency']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numerical_cols)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor
model2= Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model2', GradientBoostingRegressor(n_estimators=300,learning_rate=0.05,
    max_depth=4,
    random_state=42))
])
```

```python
model2.fit(x_train, y_train)
```

```python
y_pred2=model2.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r22=r2_score(y_test,y_pred2)
r22
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error(y_test,y_pred2)
mape
```

```python
import joblib
joblib.dump(model2,'model3.pkl')
```
